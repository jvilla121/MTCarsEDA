```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# A little magic so we can see graphs
%matplotlib inline  

# overrides matplotlib's icky style
sns.set() 
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
mtcars = pd.read_csv('data_wk2/mtcars.csv')
```

```python
mtcars
```

Converting the 0's & 1's in the 'am' category to "Automatic" & "Manual":

```python
mtcars['am'] = pd.Categorical(mtcars['am'])
```

```python
mtcars.am = mtcars.am.map({0: 'Automatic', 1:'Manual'})
```

Converting the 0's & 1's in the 'vs' category to "Vshaped" & "Inline":

```python
mtcars['vs'] = pd.Categorical(mtcars['vs'])
```

```python
mtcars.vs = mtcars.vs.map({0: 'Vshaped', 1:'Inline'})
```

Renaming columns for clarity:

```python
mtcars.rename(columns={"qsec": "Qtr Mile Time (sec)", "vs": "Engine Shape", "drat": "Rear Axle Ratio", "am": "Transmission", "wt": "Weight (1,000)"}, inplace = True)
```

## Visualizations

Getting a feel for relationships between different variables via pairplot below:

```python
sns.pairplot(mtcars)
```

## Question 1: Does the amount of cylinders affect mpg?

Per the boxplot below, not only does the amount of cylinders appear to affect mpg, but there is a wider mpg variance
among the cars with the least amount of cylinders as well.

```python
sns.boxplot(x='cyl', y='mpg',data=mtcars)
```

## Question 2: Does the weight of a car affect its' mpg?

Per the scatterplot below, it seems there is a strong correlation between mpg declining as weight of a car increases:

```python
scatter1 = mtcars.plot.scatter(x='Weight (1,000)',
                               y='mpg',
                              c='DarkBlue')
```

## Question 3: What does the distribution of hp across the dataset look like?

Per below, there is a higher skew towards cars with <=200 hp

```python
sns.distplot(mtcars['hp'])
```

## Question 4: What does the relationship between displacement & hp look like?

Per the barplot below, it seems that as displacement increases, hp does as well (301 displacement seems to be the sweet spot)

```python
f, ax = plt.subplots(figsize=(16,8))
sns.barplot(x='disp', y='hp', data=mtcars,ax=ax)
```

Correlation matrix below shows there are multiple variables with high degree of correlation amongst each other:

```python
corrmat = mtcars.corr()
f, ax = plt.subplots(figsize=(12,10)) #setting some parameters of the plot to help readability
sns.heatmap(corrmat, vmax = .8, square=True)
```
